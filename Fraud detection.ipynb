# Importing the packages we need

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
```

# Importing and making the first explanatory analysis with the dataset

```python
#Importing dataset
df = pd.read_csv('creditcard.csv')
```

```python
df.head()
```

```python
#finding out if there is some null value on the dataset
df.isnull().sum()
```

No null values were found in the dataset, so lets make a quick statistical analysis and a correlation test with the variables in the dataset

```python
#checking some stats about the data
df.describe()
```

```python
#Testing the correlation between the dataset variables
df.corr()
```

We've noticed that the strogest correlation is between the amount variable and the class. Lest's see if in a heatmap it gets easier to visualize

```python
ax=plt.figure(figsize=(30,15))
plt.title("Heatmap of the variables correlation", fontsize = 33)
df_corr=df.corr()
sb.heatmap(df_corr,linewidths=1,linecolor='white')
```

None of the variables had a strong correlation with the class (if it was fraud or not), but the variable amount had a correlation, so we are going to treat it as an independent variable and the class variable as a dependent variable.

To clarify the class variable, 0= non fraud credit cards, and 1 = fraud credid cards

```python
X=df['Amount']
y=df['Class']
```

```python
#Let's reshape the X value
x=X.values.reshape(-1,1)
x
```

```python
y
```

# Normalizing data

We are going to import the standard scaler to help us to standardise the data.
We do it cause the data are in different scales, so standardising the variables is a valluable tool to help us to work with them

```python
from sklearn.preprocessing  import StandardScaler
sc= StandardScaler()
```

```python
Sd_df = pd.DataFrame(sc.fit_transform(df), columns=df.columns)
```

# Spliting dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# Testing models

We're now going to test different methods to classify the data and check which one is better to identify credit card frauds in this dataset

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
#Fitting the values in the logistic regression
lr.fit(X_train, y_train)
```

```python
#Predicting the classes 
pred=lr.predict(X_test)
```

```python
#Checking the confusion matrix on the prediction
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```

It means that between all the data we've only had 98 fraud that were classified as trully credit cards. It mighty sound a lot but based on the size of our dataset, we could actually predict a great amount of non fraudlent operations. Let's see it on percentage.

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,pred)
score*100
```

We've gotten an accuracy of almost 100%!!
Let's see how other models work with this data.

# Decision trees

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier().fit(X_train, y_train)
```

```python
pred=clf.predict(X_test)
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
score=accuracy_score(y_test,pred)
score*100
```

The decision tree actually made a good job, scoring an outstanding accuracy, with only less than 0.2 % of accuracy than the Logistic regression. How there was only a minimum difference between the models tested so far, both of them could be used to solve the fraud problematics.

# K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
pred=knn.predict(X_test)
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
score=accuracy_score(y_test,pred)
score*100
```

The knn classifyer were also able to score great accuracy vallues making it another classifyer that we could use in this context. Let's check the last classifyer.

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
pred=gnb.predict(X_test)
```

```python
cm = confusion_matrix(y_test, pred)
cm
```

```python
score=accuracy_score(y_test,pred)
score*100
```

The last classifyer tested was also able to reach outstandig values of accuracy, so we can use it as a tool to solve this problem too
